```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/tazwar70/Databases/main/FuelConsumptionCo2.csv')
dataset = dataset.astype({'MAKE':'category','MODEL':'category','ENGINESIZE':'int64','FUELCONSUMPTION_CITY':'int64','FUELCONSUMPTION_HWY':'int64','FUELCONSUMPTION_COMB':'int64','VEHICLECLASS':'category','TRANSMISSION':'category','FUELTYPE':'category'})
dataset['MAKE'] = dataset['MAKE'].cat.codes.astype('category')
dataset['MODEL'] = dataset['MODEL'].cat.codes.astype('category')
dataset['VEHICLECLASS'] = dataset['VEHICLECLASS'].cat.codes.astype('category')
dataset['TRANSMISSION'] = dataset['TRANSMISSION'].cat.codes.astype('category')
dataset['FUELTYPE'] = dataset['FUELTYPE'].cat.codes.astype('category')
dataset.dtypes
```

```python
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
print(X)
print(y)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder',OneHotEncoder(),[1,2,3,6,7])],remainder = 'passthrough')
X = ct.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.4,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_pred),1)),1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
